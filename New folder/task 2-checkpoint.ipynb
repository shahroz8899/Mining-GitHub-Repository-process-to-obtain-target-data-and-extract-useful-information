```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from CSV files
commit_df = pd.read_csv('kubernetes_commits.csv')
issue_df = pd.read_csv('kubernetes_issues.csv')

# Task 1: Distribution of contributors' daily contributions
daily_contributions = commit_df.groupby('commit_time').size()
daily_contributions.plot(figsize=(10, 6), title='Contributors\' Daily Contributions')
plt.xlabel('Date')
plt.ylabel('Number of Commits')
plt.show()

# Task 2: Most frequent types of issues
issue_types = issue_df['issue_type'].value_counts().nlargest(10)
issue_types.plot(kind='bar', figsize=(10, 6), title='Most Frequent Types of Issues')
plt.xlabel('Issue Type')
plt.ylabel('Number of Issues')
plt.show()

# Task 3: Monthly number of commits, open issues, and closed issues
commit_df['commit_month'] = pd.to_datetime(commit_df['commit_time']).dt.to_period('M')
monthly_commits = commit_df['commit_month'].value_counts().sort_index()
monthly_open_issues = issue_df[issue_df['issue_status'] == 'open']['issue_open_time'].apply(lambda x: pd.to_datetime(x).to_period('M')).value_counts().sort_index()
monthly_closed_issues = issue_df[issue_df['issue_status'] == 'closed']['issue_close_time'].apply(lambda x: pd.to_datetime(x).to_period('M')).value_counts().sort_index()

monthly_data = pd.DataFrame({
    'Commits': monthly_commits,
    'Open Issues': monthly_open_issues,
    'Closed Issues': monthly_closed_issues
})

monthly_data.plot(kind='bar', stacked=True, figsize=(12, 8), title='Monthly Number of Commits and (Open/Closed) Issues')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the saved data from CSV files
commit_df = pd.read_csv('kubernetes_commits.csv')
issue_df = pd.read_csv('kubernetes_issues.csv')

# Convert 'commit_time' column to datetime format
commit_df['commit_time'] = pd.to_datetime(commit_df['commit_time'])

# 1) Distribution of contributors’ daily contributions
daily_contributions = commit_df.groupby('commit_time').size()
plt.figure(figsize=(12, 6))
daily_contributions.plot(kind='line')
plt.title('Distribution of Contributors’ Daily Contributions')
plt.xlabel('Date')
plt.ylabel('Number of Commits')
plt.show()

# 2) Most frequent types of issues
issue_types_count = issue_df['issue_type'].value_counts().head(10)
plt.figure(figsize=(12, 6))
issue_types_count.plot(kind='bar')
plt.title('Top 10 Most Frequent Types of Issues')
plt.xlabel('Issue Type')
plt.ylabel('Number of Issues')
plt.show()

# 3) Monthly number of commits, open issues, and closed issues
commit_df['commit_month'] = commit_df['commit_time'].apply(lambda x: x.strftime('%Y-%m'))
issue_df['issue_month'] = pd.to_datetime(issue_df['commit_time']).dt.strftime('%Y-%m')

monthly_commits = commit_df.groupby('commit_month').size()
monthly_open_issues = issue_df[issue_df['issue_status'] == 'open'].groupby('issue_month').size()
monthly_closed_issues = issue_df[issue_df['issue_status'] == 'closed'].groupby('issue_month').size()

plt.figure(figsize=(12, 6))
monthly_commits.plot(label='Commits', marker='o')
monthly_open_issues.plot(label='Open Issues', marker='o')
monthly_closed_issues.plot(label='Closed Issues', marker='o')

plt.title('Monthly Number of Commits, Open Issues, and Closed Issues')
plt.xlabel('Month')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the commit and issue datasets
commit_df = pd.read_csv('kubernetes_commits.csv')
issue_df = pd.read_csv('kubernetes_issues.csv')

# Set the style for Seaborn plots
sns.set(style="whitegrid")

# Visualization 1: Distribution of contributors' daily contributions
plt.figure(figsize=(12, 6))
sns.histplot(data=commit_df, x='commit_time', bins=30, kde=True, color='blue')
plt.title('Distribution of Contributors\' Daily Contributions')
plt.xlabel('Commit Time')
plt.ylabel('Count')
plt.show()

# Visualization 2: Most frequent types of issues
plt.figure(figsize=(10, 6))
top_issue_types = issue_df['issue_type'].value_counts().nlargest(5)
sns.barplot(x=top_issue_types.index, y=top_issue_types.values, color='green')
plt.title('Most Frequent Types of Issues')
plt.xlabel('Issue Type')
plt.ylabel('Count')
plt.show()

# Visualization 3: Monthly number of commits
commit_df['month'] = pd.to_datetime(commit_df['commit_time']).dt.to_period('M')

plt.figure(figsize=(12, 6))
monthly_commits = commit_df['month'].value_counts().sort_index()
sns.lineplot(x=monthly_commits.index.astype(str), y=monthly_commits.values, marker='o', color='purple')
plt.title('Monthly Number of Commits')
plt.xlabel('Month')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Visualization 4: Open and closed issues
plt.figure(figsize=(10, 6))
open_closed_issues = issue_df['issue_status'].value_counts()
sns.barplot(x=open_closed_issues.index, y=open_closed_issues.values, color='red')
plt.title('Open and Closed Issues')
plt.xlabel('Issue Status')
plt.ylabel('Count')
plt.show()
```
