```python
import pandas as pd
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import seaborn as sns

# Specify a font that includes CJK glyphs (Arial Unicode MS is an example)
plt.rcParams['font.family'] = 'Arial Unicode MS'

# Load the data from CSV files
commit_df = pd.read_csv('kubernetes_commits.csv')
issue_df = pd.read_csv('kubernetes_issues.csv')

# Task 1: Distribution of contributors' daily contributions
# Group commit data by committer and date
daily_contributions = commit_df.groupby(['commit_time', 'committer_id']).size().reset_index(name='contributions')

# Plot the distribution of contributors' daily contributions
plt.figure(figsize=(12, 6))
sns.lineplot(x='commit_time', y='contributions', hue='committer_id', data=daily_contributions)
plt.title('Distribution of Contributors\' Daily Contributions')
plt.xlabel('Date')
plt.ylabel('Number of Contributions')
plt.xticks(rotation=45)
plt.legend(title='Committer ID', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Task 2: Most frequent types of issues
# Group issue data by issue type and count the occurrences
issue_type_counts = issue_df['issue_type'].value_counts()

# Plot the most frequent types of issues
plt.figure(figsize=(10, 6))
issue_type_counts.plot(kind='bar')
plt.title('Most Frequent Types of Issues')
plt.xlabel('Issue Type')
plt.ylabel('Number of Occurrences')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
